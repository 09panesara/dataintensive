```python
import os
os.environ['PYSPARK_SUBMIT_ARGS'] = '--jars spark_jars/spark-xml_2.12-0.6.0.jar pyspark-shell'

from pyspark.sql.functions import explode
from pyspark.sql.functions import flatten
import csv

from pyspark import SparkContext, SparkConf
from pyspark.sql import SQLContext
conf = SparkConf().setAppName('DDI')
sc = SparkContext(conf=conf)
sqlc = SQLContext(sc)
```

```python
df = sqlc.read \
        .format("xml") \
        .option("rootTag", "drugbank") \
        .option("rowTag", "drug") \
        .load("/Users/blancabastardescliment/Desktop/courses/DIC/project/full_database.xml")
```

```python
# print the dataframe schema
df.printSchema()
```

```python
df2 = df.select(explode("drug-interactions.drug-interaction"), df["drugbank-id._VALUE"].getItem(0))
df2Flatten = df2.toDF("drugA","drugB")

df3 = df2Flatten.select(("drugA.drugbank-id"),("drugA.name"),"drugB")
interactions_df = df3.toDF("drugA_id","drugA_name","drugB_id")
interactions_df.printSchema()
interactions_df.show(10)
```

```python
interactions_df.repartition(1).write.option("header", "true").option("delimiter", "\t").csv("drug_interactions.tsv")
```
