```python
import os
import collections
import xml.etree.ElementTree as ET
import pandas as pd
import csv
```

```python
xml_file = "full_database.xml"
tree = ET.parse(xml_file)
root = tree.getroot()
```

```python
rows = list()
ns = '{http://www.drugbank.ca}'

for i, drug in enumerate(root):
    row = collections.OrderedDict()
    
    row['drugbank_id'] = drug.findtext(ns + "drugbank-id[@primary='true']")
    row['name'] = drug.findtext(ns + "name")

    for item in drug.findall("{ns}drug-interactions/{ns}drug-interaction".format(ns=ns)):

        interaction = [''.join(chid.text) for chid in item]
        interaction.append(row['drugbank_id'])

        rows.append(interaction)
```

```python
#   drug_interaction_id: id of drug A
#   name: name of drug A
#   description: interaction info of drug A with drug B
#   drugbank_id: id of drug B
interactions_df = pd.DataFrame(rows,columns=['drug_interaction_id','name','description','drugbank_id'])
```
